### Mensaje a transmitir

#### La emisión de gases de efecto invernadero ha aumentado en los últimos años.

```python
years = historical_emissions.columns[5:]
avg_emissions_per_year = historical_emissions[years].mean()

years = years[::-1] # Invertir el orden de los años para que se muestren en orden cronológico
avg_emissions_per_year = avg_emissions_per_year[::-1]

plt.figure(figsize=(10, 6))
plt.plot(years, avg_emissions_per_year, marker='o', linestyle='-', color='skyblue')

plt.title('Promedio de emisiones de gases de efecto invernadero (1990-2018)', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Emisiones promedio (toneladas)', fontsize=14)

plt.ylim(bottom=0)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Para reducir la carga cognitiva se decidió lo siguiente:

- Colores más suaves: Se utilizó una paleta de colores más suaves para que la visualización no sea tan agresiva.
- Títulos más grandes: Se aumentó el tamaño de los títulos para que sean más fáciles de leer.
- Leyenda más clara: Se cambió la leyenda para que sea más clara y se pueda entender mejor.
- Cuadrícula: Se agregó una cuadrícula para que sea más fácil seguir la línea de tiempo.
- Tamaño de la visualización: Se aumentó el tamaño de la visualización para que sea más fácil de leer.
- Limitar el rango de años: Se limitó el rango de años para que sea más fácil de leer.

#### Los países con las mayores emisiones de gases de efecto invernadero en 2018

```python
emissions_2018 = historical_emissions.groupby('Country')['2018'].sum().sort_values(ascending=False).head(11)

emissions_2018 = emissions_2018.drop('World', errors='ignore')

emissions_2018.plot(kind='barh', figsize=(12, 6), color='lightcoral')

plt.title('Top 10 países con mayores emisiones en 2018 (excluyendo "World")')
plt.xlabel('Emisiones (toneladas)')
plt.ylabel('País')

for index, value in enumerate(emissions_2018):
    plt.text(value, index, f' {value:,.0f}', va='center')

plt.grid(False)


plt.tight_layout()
plt.show()
```

Para reducir la carga cognitiva se decidió lo siguiente:

- Colores más suaves: Se utilizó una paleta de colores más suaves para que la visualización no sea tan agresiva.
- Gráfico de barras horizontal: Se utilizó un gráfico de barras horizontal para que sea más fácil de leer los nombres de los países.
- Etiqutas de los ejes: Se eliminaron las etiquetas de los ejes para que la visualización sea más limpia.

#### El cambio anual en las emisiones globales de gases de efecto invernadero

```python
total_emissions = historical_emissions[historical_emissions.columns[5:]].sum()
total_emissions = total_emissions[::-1]
annual_change = total_emissions.diff()

plt.figure(figsize=(10, 6))

colors = ['#f7be79' if value > 0 else '#ddf591' for value in annual_change]

plt.bar(annual_change.index, annual_change, color=colors)

plt.title('Cambio anual en las emisiones globales (1990-2018)')
plt.xlabel('Año')
plt.ylabel('Cambio en emisiones (toneladas)')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

- Títulos más grandes: Se aumentó el tamaño de los títulos para que sean más fáciles de leer.
- Leyenda más clara: Se cambió la leyenda para que sea más clara y se pueda entender mejor.
- Código de colores: Se utilizó un código de colores para que sea más fácil seguir la línea de tiempo.
